```python
import pandas as pd

# Read the Spotify DataFrame from the given URL
spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')

# Display the first few rows of the DataFrame
spotify.head()
```

```python
# Count the number of unique artists in the DataFrame
unique_artists_count = spotify['artist_name'].nunique()
print(f"Number of unique artists: {unique_artists_count}")

# Sort the DataFrame by the number of tracks for each artist in descending order
artist_track_counts = spotify['artist_name'].value_counts()
sorted_artist_track_counts = artist_track_counts.sort_values(ascending=False)
print("\nTop 10 artists with the most tracks:")
print(sorted_artist_track_counts.head(10))

# Indexing: Display information about Drake
drake_info = spotify[spotify['artist_name'] == 'Drake'].head(1)
print(f"\nInformation about Drake:")
print(drake_info)

# Filtering: Create a new DataFrame for tracks by Drake
drake_tracks = spotify[spotify['artist_name'] == 'Drake']

# Display the first few rows of the new DataFrame
print("\nTracks by Drake:")
print(drake_tracks.head())
```
